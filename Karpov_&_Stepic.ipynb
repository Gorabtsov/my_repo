Напишите функцию calculator, принимающую 3 аргумента: первый — операция, вторые два — числа, над которыми должна быть произведена операция.

Например: если первый аргумент +, то нужно сложить числа a и b; если —, то вычесть.

Функция должна вернуть одно число — результат выполнения операции. Если операцию выполнить не удалось — вернуть строку “Error”.

```python
# def calculator(op, a, b):
#    ...


# calculator('+', 2, 4) # -> 6
# calculator('^', 2, 4) # -> 16
# calculator('', 2, 4) # -> 'Error'
```

```python
def calculator(op, a, b):
    operations = {'+': a + b, '-': a - b, '*': a * b, '^': a ** b}
    for key, value in operations.items():
        if op == key:
            result = value
            break
        elif op == '/' and b != 0:
            result = a / b
        else:
            result = 'Error'
    return result
```

Гриффиндор — храбрость, благородство, честь;

Когтевран — ум, любознательность, творчество;

Напишите функцию sorting_hat(), которая в качестве аргумента будет принимать словарь вида:

new_students = {‘Сириус Блэк’: ‘храбрость’}, где ключ — имя ученика, значение — качество, которым он обладает. 

Функция должна вернуть 2 переменные:

sorted_students — словарь вида: ключ — имя ученика, значение — факультет, на который он был зачислен. Данные в словаре должны быть отсортированы по названию факультета и по имени ученика. 

department — словарь вида: ключ - факультет, значение — количество зачисленных на него учеников. Отсортируйте словарь по ключу. 

```python
# def sorting_hat(new_students):
#     ...
#     return sorted_students, departments

new_students = {'Сириус Блэк': 'храбрость', 'Аманда Коршун': 'любознательность',
                 'Пенелопа Вулпинголд': 'находчивость', 'Артур Поттер': 'храбрость', 'Тесая Блэк': 'ум'}
# sorting_hat(new_students) # ->
# sorted_students = {'Артур Поттер': 'Гриффиндор', 'Сириус Блэк': 'Гриффиндор', 'Аманда Коршун': 'Когтевран',
#                    'Тесая Блэк': 'Когтевран', 'Пенелопа Вулпинголд': 'Слизерин'}
# departments = {'Гриффиндор': 2, 'Когтевран': 2, 'Пуффендуй': 0, 'Слизерин': 1}
```

```python
def sorting_hat(new_students):
    facultets = {'Гриффиндор': ['храбрость', 'благородство', 'честь'],
                 'Пуффендуй': ['трудолюбие', 'верность', 'честность'],
                 'Когтевран': ['ум', 'любознательность', 'творчество'],
                 'Слизерин': ['хитрость', 'амбициозность', 'находчивость']}
    students = {}
    departments = {}

    for key_new_students, value_new_students in new_students.items():
        for key_facultets, value_facultets in facultets.items():
            if value_new_students in value_facultets:
                students[key_new_students] = key_facultets

    sorted_students = dict(sorted(students.items(), key = lambda x: (x[1], x[0])))

    for key_facultets, value_facultets in facultets.items():
        for key_sorted_students, value_sorted_students in sorted_students.items():
            if key_facultets == value_sorted_students:
                departments[key_facultets] = list(sorted_students.values()).count(value_sorted_students)
                break
            else:
                departments[key_facultets] = 0
    departments = dict(sorted(departments.items()))

    return sorted_students, departments
```

Если n чётное, то делим его на 2;
Если n нечётное, то умножаем на 3 и прибавляем 1 (получаем 3n + 1).
Напишите функцию def collatz_conjecture(), которая будет принимать натуральное число n и возвращать последовательность чисел, полученных указанными выше преобразованиями. 

```python
# def collatz_conjecture(n):
#     ...

# collatz_conjecture(8) # -> [8, 4, 2, 1]
# collatz_conjecture(3) # -> [3, 10, 5, 16, 8, 4, 2, 1]
```

```python
def collatz_conjecture(n):
    result = [n]
    while n > 1:
        if n % 2 == 0:
            n = n // 2
        else:
            n = n*3 + 1
        result.append(n)
    return result
```

```python
# def reverse_tuple(lst):
#     ...

# reverse_tuple([1, 3, 4]) # -> (4, 3, 1)
# reverse_tuple([1, 3, 4, 4, 5, 2]) # -> (2, 5, 4, 3, 1)
```

```python
def reverse_tuple(lst):
    for num in lst:
        if lst.count(num) != 1:
            lst.remove(num)
    return tuple(lst[::-1])
```

# Retention

```python
# Первый вариант:
# берется нужный временной интервал, например, день, неделя, месяц, год или любой другой.
# Считается количество пользователей, впервые воспользовавшихся приложением\продуктом в указанный период.
# Находится сколько пользователей заходили в приложение на следующий день после первого посещения приложения, на второй день, на третий и т.д.
# Часто рассматривают первые 30 дней с момента первого посещения приложения\продукта.

# Далее рассчитывают Retention по формуле: ret = user_in_day_x / all_user * 100%

# Второй вариант:
# в этом варианте рассчитывается удержание пользователей строго в конкретный день после первого посещения.
# При этом рассчитывается не по всем пользователям приложения, а отдельно для пользователей, впервые посетивших приложение в один день.
# Чаще всего рассчитывается для первого, седьмого, тридцатого дня после первого посещения приложения и называются Retention первого, седьмого и тридцатого дня соответственно.

# Рассмотрим на примере "Retention седьмого дня". Пусть 2023-01-01 приложение впервые посетили 100 пользователей,
# через семь дней 2023-01-08 приложение посетили лишь 35 из этих 100. Поэтому "Retention седьмого дня" 2023-01-08 составляет 35%.

# Далее 2023-01-02 приложение впервые посетили 80 пользователей, через семь дней 2023-01-09 приложение посетили лишь 16 из этих 80.
# Поэтому "Retention седьмого дня" 2023-01-09 составляет 20%.
```

```python
users_seven_day = new_df.query('num_day == 7').groupby('date').agg(users_seven_day = ('user', pd.Series.unique))
users_seven_day
```

```python
retention_full.head(50)
```

```python
retention_full = retention_full.iloc[7:, :]
```

```python
retention_full['retention_seven_day'] = retention_full.apply(lambda x: len(np.intersect1d(x['users_null_day'], x['users_seven_day'])) / len(x['users_null_day']), axis = 1)
```
